# 2-1.집계 함수 사용
- 데이터 프레임에서 집계함수는 열단위로 수행한다.
- 결측치인 로우는 제외하고 계산한다.

|함수명|구하는 값|
|---|---|
|size|데이터의 개수|
|count|데이터의 개수|
|mean|평균|
|median|중간값|
|min|최소값|
|max|최대값|
|sum|합계|
|prod|곱|
|std|표준편차|
|var|분산|
|quantile|위치값|


```python
import numpy as np
import pandas as pd
```

```python
df1 = pd.read_csv('data/grade.csv', encoding='euc-kr')
df1.set_index('이름', inplace = True)
df1
```

# 2-2.전체 열에 대한 집계

```python
# 총점
df1.sum()
```

```python
# 평균
# 수치형 데이터가 아닌 것들은 계산이 불가능하기 때문에
# 제외한다.
df1.mean()
```

```python
# 데이터의 개수
df1.count()
```

```python
# 특정 열의 평균을 구한다.
# 통계에 관련된 함수들을 데이터 프레임뿐만 아니라 Series도 동일하게 가지고 있다.
df1['국어'].max()
```

```python
# 행단위 집계
# 문자열 데이터는 제외된다
df1[['국어','영어','수학','과학']].mean(axis=1)
```

```python
# 각 학생별 총점과 평균을 구해 데이터 프레임에 추가한다.
# 학생별 전체 총점을 구한다.
df1['총점'] = df1[['국어','영어','수학','과학']].sum(axis=1)
df1['평균'] = round(df1[['국어','영어','수학','과학']].mean(axis=1),2)
# NaN 값는 제외하고 계산하기 때문에 평균은 생각하고 계산해야 함 
df1
```

# 2-3.그룹 통계

```python
df1 = pd.read_csv('data/city_people.csv', encoding='euc-kr')
df1
```

```python
# 도시를 기준으로 인구 최대 값을 구한다
df2 = df1[['도시','인구']]

r1 = df2.groupby('도시').max()
r1
```

```python
r2 = df1.groupby('도시').max()
r2
```

```python
# 두개 이상의 컬럼을 그룹의 기준으로 설정한다.
# 각 권역에서 년도별 최대 인구수를 구한다
df2 = df1[['연도','인구','지역']]

r2 = df2.groupby(['지역','연도']).max()
r2
```

```python
# 기준이 되는 컬럼을 데이터로 포함시키고 심다면 as_index에 False를 넣어준다.

df2 = df1[['도시','인구']]
r1 = df2.groupby('도시', as_index=False).max()

df3 = df1[['지역', '연도', '인구']]
r2 = df3.groupby(['지역','연도'], as_index=False).max()

display(r1)
display(r2)
```

# 2-4.여러 통계 값을 동시에 구하기

```python
df2 = df1[['도시','인구']]
df2
```

```python
# 각 도시별 최소, 최대, 표준편차, 총합을 구한다.
# 구할 통계에 대한 함수의 이름을 문자열로 넣어준다.

df2.groupby('도시').agg(['min','max','std','sum'])
```

```python
# 각 컬럼별 서로 다른 통계 값을 구한다.
df2 = df1[['도시','인구','지역']]

a1 = {
    '도시' : ['count'],
    '인구' : ['sum','max']
}
df2.groupby('지역').agg(a1)
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression , SGDRegressor # LinearRegression : 공식 , SGDRegressor 학습
from sklearn.neural_network import MLPRegressor # (딥러닝)학습
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
matplotlib.rcParams['font.family']='Malgun Gothic'
matplotlib.rcParams['axes.unicode_minus'] = False

from sklearn.metrics import r2_score # 선형회귀 결정계수

from sklearn.datasets import load_boston, load_iris, fetch_california_housing
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler
from sklearn.pipeline import make_pipeline

from statsmodels.stats.outliers_influence import variance_inflation_factor # VIF
```

```python
carDF = pd.read_csv('datas/cars.csv')
carDF.columns = ['speed','dist']
carDF
```

```python
x_data = carDF[['speed']]
y_data = carDF['dist']
```

# SGD Regressor 경사하강법

```python
# max_iter = 1000 >> 기본 학습 회수 1000회
# alpha = 0.0001  >> learning rate
# early_stopping = False >> True로 바꾸면 cost 값이 어느 정도 변화가 없으면 종료

modelSGD = SGDRegressor(max_iter=200, alpha=0.001, early_stopping=True, verbose=1)
modelSGD.fit(x_data,y_data)
```

```python
display(modelSGD.coef_)
display(modelSGD.intercept_)
```

```python
y_predict = modelSGD.predict(x_data)
plt.scatter(x_data,y_data)
plt.plot(x_data,y_predict,'r')
plt.show()
```

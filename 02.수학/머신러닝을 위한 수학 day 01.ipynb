# 의사역행렬

```python
a = np.array([[1,1],
             [2,1],
             [3,1]])
y = np.array([[2], 
              [4],
              [6.4]])
y = np.array([2,4,6.4]).reshape(3, 1)
y = np.array([2,4,6.4]).reshape(-1, 1)
y = np.array([[2,4,6.4]]).T

pinv  = np.linalg.pinv(a)
print(pinv)

np.matmul(pinv, a)

W = np.matmul(pinv, y)
print(W)
```

```python
#print(a)
mse = np.mean((np.matmul(a, W) - y)**2)
print(mse)
```

```python
def f(x) :
    return 4*x + 7

x = 0
dx = 0.000000001
(f(x+dx) - f(x) ) /  dx

def df(x) :
    dx = 0.000001
    return (f(x+dx) - f(x) ) /  dx

df(1)
```

```python
def f2(x) :
    return 4*x**2 +   20*x - 5   #    8x+20

def df2(x) :
    dx = 0.000001
    return (f2(x+dx) - f2(x) ) /  dx

f2(5), df2(5)
```

```python
def df_g(f, x) :
    dx = 0.000001
    return (f(x+dx) - f(x) ) /  dx  

def fun1(x) : 
    return 5*x + 7

df_g(fun1, 6)
```

```python
def sig(x) :
    return  1 / (1 + np.exp(-x))
df_g(sig, -10000)
```

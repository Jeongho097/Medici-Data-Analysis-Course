```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 폰트 설정 - 윈도우 
plt.rcParams['font.family'] = 'Malgun Gothic'

# 글자 크기
plt.rcParams['font.size'] = 15

# 폰트 설정 시 - 기호 깨는 거 방지하기
plt.rcParams['axes.unicode_minus'] = False

# 그래프의 크기
plt.rcParams['figure.figsize'] = 15,7
```

```python
# 2차원으로 3개씩 들어가있는 리스트가 총 100개
x = np.random.randn(100,3) 

# 2021-01-01 부터 100일치 날짜를 가져온다
index = pd.date_range('2021/01/01',periods=100) 

df1 = pd.DataFrame(x,index=index, columns=['A','B','C'])
df1
```

# 1.선 그래프

```python
# 각 컬럼의 데이터가 그래프 하나가 된다.
# 범례도 자동으로 나타나며 범례에 표시되는 이름은 컬럼의 이름으로 설정된다.
df1.plot()

plt.title('선 그래프')
plt.xlabel('시간')
plt.ylabel('Data')

plt.show()
```

# 2.다른 그래프 그리기
- plot 함수에 kind라는 매개변수에 그래프의 종류를 지정할 수 있다
- bar, pie, hist, kde, box, scatter, area

```python
iris = pd.read_csv('data/iris.csv')
iris.columns = ['꽃받침길이','꽃받침너비','꽃잎길이','꽃잎너비','품종']
iris
```

```python
# Series 하나의 데이터
iris['꽃받침길이'][:20].plot(kind='bar')
plt.title('꽃받침길이 막대 그래프')
plt.show()
```

- plot 함수의 kind에 설정하는 문자열과 동일한 이름의 함수도 제공한다.

```python
# 품좀을 제외한 나머지를 가지고 그래프를 그려본다
# 하나의 로우를 추출하여 로우 내의 컬럼별 막대 그래프를 그리고 그것을 하나의 그룹으로 묶어 준다
df2 = iris.drop('품종',axis=1)
df2.head(20).plot.bar()
```

```python
# 품종별 컬럼의 평균을 구한다.
a1 = iris.groupby('품종').mean()

a1.plot.bar()
plt.show()
```

```python
a1 = iris['품종'].value_counts()
# 매개변수에 들어가는 것들은 matplotlib에서 제공하는 pie plot과 동일하다
a1.plot.pie(autopct='%.2f%%',startangle=90)
plt.show()
```

```python
iris.plot.hist()
plt.show()
```

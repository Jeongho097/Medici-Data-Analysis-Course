# 4.종가를 시각화 한다.
- 하위 10% 이하, 상위 90% 이상을 다른 색상으로 표시한다.
- 10% 지점, 90% 지점에 기준선을 표시한다.
- x 축은 일자, y축은 종가로 표시한다.

```python
cel_close = cel['Close']
cel_close
```

```python
# 시각화
# 데이터 프레임이나 시리즈로 선그래프를 그리면
# index가 x축으로 지정되고 y축이 값으로 지정된다.
cel_close.plot(color='black')
plt.show()
```

# 5.특정 범위를 벗어나는 그래프의 색을 다르게 표시한다.
- 10%, 90% 

```python
# 방법 1
value_10 = cel_close.quantile(0.1)
value_90 = cel_close.quantile(0.9)

display(value_10)
display(value_90)
```

```python
# 방법 2
a1 = cel_close.describe(percentiles=[0.1,0.9])

value_10 = a1['10%']
value_90 = a1['90%']

display(value_10)
display(value_90)
```

```python
# 10% 이하의 값들
a2 = cel_close <= value_10

# 90% 이상의 값들
a3 = cel_close >= value_90

# 둘을 합친다.
a4 = a2 | a3

a4
```

```python
# 10% 이하, 90% 이상 값
cel_top_bottom = cel_close[a4]
cel_top_bottom
```

```python
# 시각화
# style : 선의 모양. ' '을 주면 선의 모양은 빈칸이 된다. 
# marker : 마커의 모양
# ms : 마커의 크기
# color : 색상

cel_top_bottom.plot(style=' ', marker='o', ms=4, color='darkgray')
plt.show()
```

# 6.기준선을 그린다

```python
# x축의 좌표로 사용할 최소 날짜 최대 날짜를 추출한다.
index_min = cel_close.index[0]
index_max = cel_close.index[-1]

display(index_min)
display(index_max)
```

```python
# 가로선을 그린다.
plt.hlines(xmin=index_min, xmax=index_max,
           y=[value_10,value_90],
          color='red')
```

# 7.3가지를 합친다.

```python
cel_close.plot(color='black')

cel_top_bottom.plot(style=' ', marker='o', ms=4, color='skyblue')

plt.hlines(xmin=index_min, xmax=index_max,
           y=[value_10,value_90],
          color='red')

plt.show()
```

# 9.첨도
- 첨도는 히스토그램에서 분포의 꼬리부분의 길이와 중앙 부분의 뾰족함에대한 정보를 제공하는 통계량이다.
- 첨도가 0보다 크면 긴 꼬리를 갖고 분포가 중앙 부분에 덜 분포되어 있는 것으로 해석된다.
image.png

```python
cel_pct_change.kurt()
```

# 10.왜도
- 분포의 비대칭 정도를 나타내는 값
- 분포가 완전 대칭인 경우에는 왜도가 0이다
- 왼쪽으로 긴 꼬리를 가지면 음수이고, 오른쪽으로 긴 꼬리를 가지면 양수
- 비대칭정도가 심할수록 왜도의 절대값이 증가한다.

image.png

```python
cel_pct_change.skew()
```

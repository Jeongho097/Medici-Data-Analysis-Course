- DataFrame의 커럼 하나는 하나의 Series로 구성되어 있으며 다수의 Series들을 모아 하나의 DataFrame으로 관리한다.
- DataFrame은 데이터의 특성 만큼의 Series로 구성된다.

### DataFrame 생성
- 2차원 리스트
- 리스트를 가지고 있는 딕셔너리
- 딕셔너리를 가지고 있는 리스트
- 외부 파일(csv, excel, spss 등)
- 웹 문서의 table tag
- 데이터 베이스의 테이블
- 기타 등등

```python
import numpy as np
import pandas as pd
```

```python
# 2차원 리스트를 통한 생성
# 내부의 각 리스트의 인덱스가 일치하는 것 끼리 모아 Series 만든다.
# 데이터 개수가 다르게 되어 있을 경우에는 비어있는 곳은 결측치로 채워진다.

list1 = [
    [1, '남자', 90,80,70],
    [2, '여자', 91,81,71],
    [3, '남자', 92,82,72],
    [4, '여자', 93,83,73],
    [5, '남자', 94,94],
    [6, '여자', 95,85,75,65]
]

df1 = pd.DataFrame(list1)
df1
```

```python
# 컬럼 이름에 사용한 이름
column_list = ['번호','성별','국어','영어','수학','과학']
index_list = ['학생1','학생2','학생3','학생4','학생5','학생6']
```

```python
# 생성되어 있는 데이터 프레임에 인덱스와 컬럼 이름을 설정한다.
df1.columns = column_list
df1.index = index_list

df1
```

```python
# 생성 시 인덱스와 컬럼 이름을 지정한다.
df2 = pd.DataFrame(list1,
                   columns = column_list, 
                   index=index_list)
df2
```

```python
# 리스트를 가지고 있는 딕셔너리를 통한 생성
# 딕셔너리에 지정한 이름이 컬럼의 이름으로 결정된다.
# 데이터의 수가 다르게 되어 있으면 오류가 발생한다.
data1 = {
    '번호' : [1,2,3,4],
    '성별' : ['남자', '여자', '남자', '여자'],
    '국어' : [90, 91, 92, 93],
    '영어' : [80, 81, 82, 83],
    '수학' : [70, 71, 72, 73]
#     '과학' : [60, 61, 62, 63, 64] 
#     '한국사' : [50, 51, 52]      
}

df3 = pd.DataFrame(data1)
df3
```

```python
# 딕셔너리를 가지고 있는 리스트
# 각 딕셔너리에 이름이 똑같은 것들까지 하나의 Series로 구성한다
list3 = [
    {'번호' : 1, '이름' : '홍길동', '국어' : 100},
    {'번호' : 2, '이름' : '김길동', '국어' : 90},
    {'번호' : 3, '이름' : '이길동', '국어' : 80},
    {'번호' : 4, '이름' : '촤길동', '국어' : 70},
    {'이름' : '항길동', '국어' : 60, '번호' : 5},
    {'번호' : 6, '이름' : '박길동', '귝어' : 105},
    {'번호' : 1, '이름' : '홍길동', '국어' : 100, '영어' : 80}
]

df4 = pd.DataFrame(list3)
df4
```

```python
# csv 파일
# 데이터를 쉼표(,)로 구분한 데이터 양식
# 데이터프레임 생성 시 Encoding 오류가 발생하거나 한글 데이터가 깨지는 경우
# pandas는 기본 Encoding이 UTF-8로 설정되어 있다.
# 한국에서는 엑셀을 통해 CSV 파일을 생성 시 euc-kr로 되어 있다.
# csv 파일을 통해서 생성 시 Encoding 오류가 나거나 한글이 깨져서 나오면
# Encoding을 euck-kr로 설정해서 시도해본다

# 첫 번째 줄 데이터가 컬럼의 이름으로 결정된다

df1 = pd.read_csv('data/grade.csv', encoding='euc-kr')

df1
```

# 4.원하는 데이터를 가져오기

```python
# 컬럼을 가져온다 (Series로 반환된다.)
df1['Survived']
```

```python
# 여러 컬럼들을 가져온다
df1[['Survived','Name','Cabin','Embarked']]
```

```python
# 범위형 데이터에 대한 요약 통계 정보
# count : 결측치 를 제외한 데이터의 개수
# mean : 결측치를 제외한 평균
# std : 결측치를 제외한 표준편차
# min : 결측치를 제외한 최솟값
# 25% : 최대 최소 범위 안에  25% 위치의 값(존재하지 않을 수도 있다.)
# 50% : 최대 최소 범위 안에  50% 위치의 값(존재하지 않을 수도 있다.)
# 75% : 최대 최소 범위 안에  75% 위치의 값(존재하지 않을 수도 있다.)
# max : 결측치를 제외한 최댓값

# min과 max를 통해 정상 범위에 벗어나는 데이터가 있는지 확인한다.

df1[['Age','Fare']].describe()
```

```python
# 레이블형 데이터
# count : 결측치를 제외한 데이터의 개수
# unique : 값의 종류
# top : 가장 많이 저장되어 있는 값
# freq : 가장 많이 저장되어 있는 값이 몇 개가 저장되어 있는지

# 요약 통계 정보는 숫자 컬럼인 경우에는 범위형 데이터로 취급해 보여주게 되고
# 문자열이라고 하더라도 어떠한 데이터가 저장되어 있는지 알 수 없기 때문에 이상치 존재 여부를 확인하기가 어렵다.

df1['Embarked'].describe()
```

```python
# value_counts : Series 내에 어떠한 값이 몇 개씩 저장되어 있는지 결과를 반환해준다.
display(df1['Pclass'].value_counts())
display(df1['Embarked'].value_counts())
display(df1['Embarked'].value_counts().index)
```

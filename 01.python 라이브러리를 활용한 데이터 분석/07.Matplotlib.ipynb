# 1.Matplotlib
- Matplotlib 파이썬에서 자료를 시각화하는 패키지
- 라인 플롯, 스캐터 플롯, 컨투어 플롯, 서피트 플롯, 바 차트, 히스토그램, 박스 플롯 등등 다양한 그래프를 그릴 수 있다
- https://matplotlib.org/stable/gallery/index.html

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np

# 그래프를 그리기 위한 기본 설정

# 폰트 설정 - 윈도우 
plt.rcParams['font.family'] = 'Malgun Gothic'

# 글자 크기
plt.rcParams['font.size'] = 15

# 폰트 설정 시 - 기호 깨는 거 방지하기
plt.rcParams['axes.unicode_minus'] = False

# 그래프의 크기
plt.rcParams['figure.figsize'] = 15,7
```

# 2.선 그래프
- 가장 간단한 그래프이다.
- 데이터가 시간, 순서 등에 따라 어떻게 변화되는지를 보여주기 위해 사용된다.

```python
x = [1,2,3]
y = [2,3,1]

plt.plot(x,y, 'bo-.')

plt.title('선 그래프')

# 각 축의 값의 종류와 동일한 것으로 설정해야 한다.
# 선 그래프에 어울리는 데이터틑 숫자인 경우이기 때문에 숫자로 설정하면 된다.
plt.xticks([1,2,3], ['가','나','다']) 
plt.yticks([2,3,1], ['중','상','하']) 

# 각 축의 이름
plt.xlabel('x축')
plt.ylabel('y축')

plt.show()
```

# 3.막대 그래프
- 데이터들의 양을 비교하기 위해 사용한다.

```python
x = [1,2,3]
y = [2,3,1]

plt.bar(x,y)

plt.title('막대 그래프')

# 각 축의 값의 종류와 동일한 것으로 설정해야 한다.
# 선 그래프에 어울리는 데이터틑 숫자인 경우이기 때문에 숫자로 설정하면 된다.
plt.xticks([1,2,3], ['가','나','다']) 
plt.yticks([1,2,3], ['하','중','상']) 

# 각 축의 이름
plt.xlabel('x축')
plt.ylabel('y축')

plt.show()
```

```python
x = [1, 2, 3]
y = [2, 3, 1]

plt.title('가로 막대 그래프')
plt.barh(x,y)

# 각 축의 값의 종류와 동일한 것으로 설정해야 한다.  
# 선 그래프에 어울리는 데이터는 숫자인 경우이기 때문에 숫자로 설정하면 된다.
plt.yticks([1, 2, 3], ['가', '나', '다'])
plt.xticks([1, 2, 3], ['하', '중', '상'])

#  각 축의 이름
plt.xlabel('X축')
plt.ylabel('Y축')

plt.show()
```

# 4.스템 플롯 
- 막대 그래프와 유사하지만 폭이 없는 그래프이다.
- 이상 확률 함수나 자기 상관 관계를 묘사할 때 사용한다.

```python
x = np.linspace(0.1,2*np.pi,10)
y = np.cos(x)

plt.stem(x,y,'-.')
plt.title('스템플롯')
plt.show()
```

# 5.파이차트
- 카테고리 별 값의 상대적인 비교를 해야 할 때
- 막대 그래프는 양을 비교하는 것이고 파이차트는 비율을 비교한다.

```python
values = [15,30,40,10]
labels = ['값1','값2','값3','값4']
colors = ['yellow', 'lightgreen', 'lightskyblue','lightcoral']

# 조각의 튀어나옴 정도
explode = [0,0.1,0,0]

# 첫번째 : 값, labels = : 파이 조각의 이름, colors : 파이의 색상
# explode : 파이별로 튀어나옴 정도, shadow : 그림자 표시 여부
# startangle : 원의 시작 각도 값('기본이 0도 >> 3시 방향', 시계 반대 방향)
# counterclock : 원의 방향을 시계 반대 방향으로 할 것인가, 기본이 True
# autopct  : 파이에 설정하는 비율 값 양식, 설정하지 않으면 나타나지 않는다.

plt.pie(values, labels=labels,
       colors = colors,
       explode=explode,
       shadow=True,
       startangle=90,
       counterclock=False,
       autopct='%.2f%%'
)
plt.title('파이차트')
plt.show()
```

# 9.제목과 각 축의 글자 설정

```python
option1 = {'family' : 'NanumPen', 'size' : 24, 'color' : 'black'}
option2 = {'family' : 'NanumBarunpen', 'size' : 18, 'color' : 'darkred'}
option3 = {'family' : 'NanumBarunGothic', 'size' : 12, 'color' : 'blue'}

plt.plot([10,20,30,40,50],[-10, 1, 4, 9, 16])
plt.title('선 그래프', fontdict = option1)

# x축의 이름
plt.xlabel('x축', fontdict = option2)
# y축의 이름
plt.ylabel('y축', fontdict = option3)

plt.show()
```

# 10.그래프 스타일 지정

```python
plt.plot([10,20,30,40,50],[-10, 1, 4, 9, 16], 'rs--')
plt.title('선 그래프')

plt.show()
```

- 그래프의 스타일
 - '색상 마커종류 선종류' 형태로 작성한다. 
   - (ex 'rs--')
   - 이런식으로 하면 base color만 적용가능
   

- 색상
 - https://matplotlib.org/stable/gallery/color/named_colors.html#sphx-glr-gallery-color-named-colors-py


- 마커
 - **. , o v ^ < > 1 2 3 4 s p * h H + D d**
 
- 선 스타일
|문자|의미|
|---|---|
| - |실선|
| -- |dash|
| -. |점선과 dash|
| : | 점선|

```python
plt.plot([10,20,30,40,50],[-10, 1, 4, 9, 16], 'r^-.')
plt.title('선 그래프')

plt.show()
```

# 11.상세 스타일

- c : 선 색상
- lw : 선 굵기
- ls : 선 스타일
- marker : 마커 종류
- ms : 마커 크기
- mec : 마커 색상
- mew : 마커 선 굵기
- mfc : 마커 내부 색상

```python
plt.plot([10,20,30,40,50],[-10, 1, 4, 9, 16],  
         c='b', lw=5, ls = ':', 
         marker='o', ms=15, mec='g', mew=5, mfc='r')
plt.title('선 그래프')

plt.show()
```

# 14.범례

```python
# np.linspace(시작값, 종료값, 숫자 게수) : 시작값에서부터 종료값까지 총 숫자 개수만큼
# 숫자를 가지고 있는 ndarray를 생성한다.
x = np.linspace(-np.pi,np.pi,256)
y1 = np.cos(x)
y2 = np.sin(x)

plt.plot(x,y1, ls='--', label='cos')
plt.plot(x,y2, ls=':', label='sin')
plt.title('범례 표시하기')

# 각 그래프의 label 값을 활용하여 범례를 표시한다.

# loc : 범례의 위치, 생략시 자동
# 0(best, 기본), 1(우상단), 2(좌상단), 3(좌하단), 4(우하단), 5(우측)
# 6(중앙좌측), 7(중앙우측), 8(하단중앙), 9(상단중앙), 10( 정중앙)

plt.legend(loc = 10)

plt.show()
```

# 1.Series

# 2.Series 생성
- Series는 list, tuple, ndarray로 생성할 수 있다.
- list, tuple인 경우에는 ndarray로 생성한 다음 관리한다.
- ndarray인 경우에는 해당 행렬 데이터를 관리한다.

```python
import numpy as np
import pandas as pd
```

```python
s1 = pd.Series([10,20,30,40,50])
print(s1)
display(s1)
```

# 3.Series가 관리하는 데이터에 접근하기

```python
print(s1[0]) # 이건 인덱스를 지칭하는 거임
print(s1[1])
print(s1[2])
```

```python
array1 = s1.values
print(type(array1))
print(array1)
```

```python
array2 = s1.index
print(type(array2))
print(array2)
```

# 4.index 지정하기
- Series 생성 시 index를 지정하지 않으면 0부터 1씩 증가되는 값으로 지정된다.
- index를 지정하고 싶다면 생성 시 지정하거나 생성 후 필요할 때 지정할 수 있다.

```python
s1 = pd.Series([1000,2000,3000])
s1
```

```python
# 생성한 Series에 인덱스를 변경한다.
# 새롭게 설정하는 인덱스의 개수는 Series의 전체 행의 개수와 일치해야 한다.
s1.index = ['값1','값2','값3']
s1
```

```python
s2 = pd.Series([1000,2000,3000], index=['값1','값2','값3'])
s2
```

```python
# 기본 index
s1 = pd.Series([10,20,30,40,50])
print(s1)

s2 = pd.Series([1000,2000,3000], index=['값1','값2','값3'])
print(s2)
```

```python
# index가 숫자로 되어 있을 경우 숫자를 지정하면 순서가 아닌
# index 값을 통해 접근한다.
print(s1[0])
print(s1[1])
print(s2['값1'])
print(s2[0])
```

```python
s2 = pd.Series([10,20,30,40,50], index = [3,0,4,1,1])
s2
```

```python
s2[1]
```

```python
# index를 지정했더라도 숫자로 구성되어 있으면 []의 숫자는 순서가 아닌
# index 값을 의미한다.
print(s2[0])
print(s2[3])
print(s2[4])
```

```python
# 인덱스가 문자열인 Series
a1 = ['값1','값2','값3','값4','값5']
s3 = pd.Series([10,20,30,40,50], index = a1)
s3
```

```python
print(s3['값1'])
print(s3['값2'])
print(s3['값3'])
```

```python
# index가 숫자가 아닌 Series에 숫자를 통해 접근을 하게 되면 순서에 해당한다.
# 0부터 1씩 증가한다.
```

```python
print(s3[0])
print(s3['값1'])
print(s3[1])
print(s3['값2'])
print(s3[2])
print(s3['값3'])
```

# 5.Series의 연산
- Series는 내부적으로 1차원 ndarray를 관리하기 때문에 연산 등의 작업은 ndarray와 동일하다.

```python
s1 = pd.Series([10, 20, 30, 40, 50])
s2 = pd.Series([1, 2, 3, 4, 5])
```

# 6.Series와 Series의 연산
- 각 Series의 데이터를 1 : 1 매칭하여 연산을 한다.
- 개수가 다르면 NaN 값을 결과값으로 설정한다.

```python
s10 = s1 + s2
s11 = s1 - s2
s12 = s1 * s2
s13 = s1 / s2
print(s10)
print(s11)
print(s12)
print(s13)
```

# 7.개수가 다른 Series의 연산
- Pandas는 연산이 성립하지 않는 상황이라면 오류가 발생하는 것이 아니라
- NaN이라는 값을 결과로 결정한다
- NaN : Not a Number. 결측치를 의미하는 값이다.
- 결측치 : 측정이 되지 않는 값, 값이 비어있는 공간

```python
a100 = pd.Series([1,2,3])
a200 = pd.Series([10,20,30,40])

a100 + a200
```

```python
# Series와 특정 값의 계산
s20 = s1 + 10
s21 = s1 - 10
s22 = s1 * 10
s23 = s1 / 10

print(s20)
print(s21)
print(s22)
print(s23)
```

```python
r1 = s1 > 30
r1
```

# 8.슬라이싱

```python
s1 = pd.Series([10,20,30,40,50],index = ['a1','a2','a3','a4','a5'])
s1
```

```python
s1['a1' : 'a3']
```

```python
s4 = s1[['a1','a3','a5']]
s4
```

```python
# 순서 인덱스
s5 = s1[[0,2,4]]
s5
```

```python
# True난 False를 넣으면 True 자리에 해당하는 것만 추출한다.
s6 = s1[[True,False,True,False,True]]
s6
```

```python
# 조건에 따라 가져온다
s7 = s1 > 30
s7
```

```python
s8 = s1[s7]
s8
```

```python
s9 = s1[s1 > 30]
s9
```

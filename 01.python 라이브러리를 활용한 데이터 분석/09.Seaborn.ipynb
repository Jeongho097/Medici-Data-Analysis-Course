```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 폰트 설정 - 윈도우 
plt.rcParams['font.family'] = 'Malgun Gothic'

# 글자 크기
plt.rcParams['font.size'] = 15

# 폰트 설정 시 - 기호 깨는 거 방지하기
plt.rcParams['axes.unicode_minus'] = False

# 그래프의 크기
plt.rcParams['figure.figsize'] = 15,7
```

# 1.1차원 분포 플롯
- 데이터의 분포를 확인하기 위해서 산점도(scatter)를 사용한다.
- 산점도는 x축과 y축이 지정되어야 하기 때문에 1차원 데이터는 표현할 수 없다
- 이에 seaborn에서 1차원 데이터의 분포를 볼 수 있는 것을 제공한다

```python
iris = pd.read_csv('data/iris.csv')
iris.columns = ['꽃받침길이','꽃받침너비','꽃잎길이','꽃잎너비','품종']
iris
```

```python
# Rug plot
# 1차원 데이터에 저장되어 있는 값들의 종류를 확인할 수 있다.
# 각 값들이 얼마나 저장되어 있는지 확인할 수 있다.
sns.rugplot(iris['꽃잎길이'])
plt.title('꽃잎 길이에 대한 Rug plot')
plt.show()
```

```python
# kde : 커널 함수를 이용해 밀도를 표현한다,
# 히스토그램을 통해 구간별 값의 개수를 파악할 수 있다,
# 히스토그램은 막대 그래프로 표현하지만. kde는 선 그래프로 각 값의 개수를 파악할 수 있다.
sns.kdeplot(iris['꽃잎길이'])
plt.title('kde plot')
plt.show()
```

```python
# distplot : 러그아 커널 밀도를 같이 표현하는 함수
# 표시되는 믹대그래프는 각 구간별 값의 종류 수를 표시한다.

sns.displot(iris['꽃잎길이'], kde=True,rug=True)
plt.title('dis plot')
plt.show()
```

# 2.count plot
- 특정 컬럼의 데이터의 개수를 막대 그래프로 표현한 그래프이다.
- 특정된 컬럼의 데이터의 각 종류 별로 몇 개씩 저장되아 있는지 확인할 수 있는 그래프이다.
- 간단하게 그릴 수 있다는 장점이 있고 matplotlib과 데이터프레임을 통해 그래프를 그릴 때 오류가 발생하는 부분을 해결한 그래프

```python
sns.countplot(data=iris, x = '품종')
plt.title('count plot')
plt.show()
```

# 3.다차원 데이터 
- 데이터 변수가 여러 개인 다차원 데이터를 표현할 수 있다.
- 분석하고자 하는 데이터가 모두 범위 값인 경우
- 분석하고자 하는 데이터가 모두 카테고리 값인 경우
- 분석하고자 하는 데이터가 모두 범위형과 카테고리형이 섞여 있는 경우

# 4.2차원 실수형 데이터

```python
# joint plot : 산점도와 히스토그램을 같이 표현한다.
sns.jointplot(x='꽃받침길이',y='꽃받침너비',data=iris)

# 제목이 이상하게 나오는 경우 suptitle을 사용
plt.suptitle('joint plot', y = 1.02)
plt.show()
```

```python
# kind에 kde를 넣어주면 kde 커널 밀도 히스토그램을 그릴 수 있다.
sns.jointplot(x='꽃받침길이',y='꽃받침너비',data=iris,
             kind='kde')
plt.suptitle('joint plot의 kde 밀도 히스토그램',y=1)
plt.show()
```

# 5.Pair plot
- 산점도의 경우에는 2차원 데이터, 즉 컬럼 두개를 지정하여 표현한다.
- 만약 데이터 프레임의 컬럼이 다수인 경우 산점도를 그리려면 개발자가 꽤 많은 그래프를 그려야한다.
- Pair plot은 데이터 프레임의 각 컬럼들을 조합한 산점도를 모두 표현해준다.
- 여기에 히스토그램도 같이 표현해 준다.

```python
sns.pairplot(iris)
plt.suptitle('pair plot', y=1.02)
plt.show()
```

```python
# hue : 결과 데이터에 해당하는 컬럼을 지정한다.
# 지정된 컬럼의 값이 똑같은 것 끼리 같은 색으로 표시한다.
sns.pairplot(iris,hue='품종')
plt.suptitle('pair plot', y=1.02)
plt.show()
```

# 6.Heatmap
- 컬럼간의 상관 관계나 데이터의 양을 색상으로 표현해 확한하는 그래프

```python
flights = sns.load_dataset('flights')
flights
```

```python
# pivot 테이블을 구한다.
# 컬럼 하나를 x축, 하나를 y축으로 지정하여 3번째로 지정된 것을 값으로 하는 테이블을 의미한다.
# 로우 하나의 데이터를 테이블 형태로 만든 테이블
df = flights.pivot('month','year','passengers')
df
```

```python
sns.heatmap(df)
plt.title('heatmap',)
plt.show()
```

```python
# 상간관계를 heatmap으로 표현한다.
# corr() 상관관계를 나타내는 함수

sns.heatmap(iris.corr(), annot=True, 
            linewidths=1, cmap='Blues')

# https://seaborn.pydata.org/tutorial/color_palettes.html
# cmap : 색상 팔레트

plt.title('heatmap')
plt.show()
```
